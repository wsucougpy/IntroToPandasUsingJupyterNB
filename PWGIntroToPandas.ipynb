# Pandas

```python
import pandas as pd
```

```python
#We are going to read the data from a .csv or "comma-separated-value" file
#data = pd.read_csv("filename.csv")
#I am using a file borrowed from support.spatialkey.com
df = pd.read_csv("TechCrunchcontinentalUSA.csv")

#####df = pd.DataFrame(data)
```

```python
#We can see the data we've just read by asking it to print out.
#print(df)

#or simply declaring:
df
```

```python
#Or we can look at just the top few rows to get a sense of what the data looks like.
df.head()
```

```python
#We can even look at a specific number of rows at the top or bottom of the list.
df.head(8)
```

```python
df.tail(10)
```

```python
#We can do all these same commands with data read from an excel sheet.
#Here is sample data borrowed from contextures.com:

xdata = pd.read_excel("SampleData.xlsx")
xdata.head()
```

```python
#Let's go back to the data from the csv file.
#We can single out the columns, rows, or labels, or even specific cells...

#df.columns   #displays headers of columns

#df['raisedAmt']    #displays everything in column labeled raisedAmt
#df.raisedAmt       #does the same thing

df.iloc[2:5]      #displays rows 2,3,4 (since Python doesn't make it to the last value)



```

```python
df.iloc[2,5]      #displays the cell in row 2 and column 5
```

```python
#Here is a for loop to print the contents of each row one by one:
for i, row in df.iterrows():
    print(i, row)
```

```python
#...or we can select certain pieces of the rows that we want to focus on:
for i, row in df.iterrows():
    print(i, row[['company','raisedAmt','fundedDate']])
```

```python
#Alternatively, we can pull out rows that only contain certain values:

df.loc[df.company == "LifeLock"]
```

```python
#Here is a slightly different way to do the same type of search
df.loc[df['company']=='Facebook']
```

```python
#How about searching with multiple criteria?

df.loc[(df['company']=='Facebook') & (df['raisedAmt']>20000000)]
```

```python
#Here is a way to do a 'find and replace'

#df.loc[df['company'] == 'Facebook', 'company'] = 'Bookface'
df.loc[df['company'] == 'Bookface', 'company'] = 'Facebook'
df.head(25)
```

```python
#df.loc[df['company'] == 'Facebook', 'company'] = 'Bookface'
#What we really just did is a conditional statement, like an IF statement
#df.loc[   IF THIS PART IS TRUE,    THEN TO THIS FIELD  ]    =   DO THIS

#Here's another example:
df.loc[df['raisedAmt'] > 30000000, 'category'] = 'SUPER!!'
df
```

```python
#We can get descriptive statistics for each column (and interpret them carefully!)
df.describe()
```

```python
#The order of the data can be changed by sorting according to a heading

df.sort_values('raisedAmt')
```

```python
#Sorting can ...theoretically ... be done with multiple categories as well
df.sort_values(['raisedAmt','city'],ascending = [True,True])
```

```python
#If we want, we can manipulate the data and create new categories/columns:

df['halfRaised'] = df.raisedAmt/2.0
df.head()
```

```python
#It could be helpful to reduce some of the clutter from our columns
df2 = df.drop(columns = ['permalink','city','raisedCurrency'])

df2.head()
```

```python
df2 = df2.sort_values('raisedAmt')
df2 = df2.reset_index(drop=True)
df2.head()
```

```python
#Notice that the old df still exists...
df.head()
```

```python
#We can write this version of the data out to its own csv file:

df2.to_csv('TechCrunchModified.csv')

#Similarly, we can write to an excel file with
#df2.to_excel('Filename.xlsx')
```
